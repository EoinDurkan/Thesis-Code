```python
import numpy as np
import xarray as xr

import cartopy.crs as ccrs
import cartopy.feature as cfeature

import matplotlib.pyplot as plt
```

```python
ATHLAT, ATHLON = [53.5, -8.]
LATLON = np.array([[53.428, -6.241],[51.847, -8.486],[54.228, -10.007],[51.938, -10.241],[55.372, -7.339],[53.537, -7.362]])
DeltaLat = 2.5
DeltaLon = 5
```

```python
LON_bounds = np.arange(ATHLON-DeltaLon, ATHLON+DeltaLon+0.01, 1)
LAT_bounds = np.arange(ATHLAT-DeltaLat, ATHLAT+DeltaLat+0.01, 1)

LON, LAT = np.meshgrid(LON_bounds, LAT_bounds)
```

```python
fig = plt.figure(figsize=[8,8])
ax = plt.axes(projection=ccrs.Mercator())
ax.coastlines(resolution='50m', color='black', linewidth=1.5, zorder=5)

for i in range(6):
    ax.plot(LATLON[i,1], LATLON[i,0], 'g*', markersize=20, transform=ccrs.PlateCarree(), zorder=10)

ax.pcolormesh(LON, LAT, np.zeros_like(LON)+100000, cmap = 'Greys', transform=ccrs.PlateCarree(), edgecolors='k', linewidths=0.1)

ax.set_extent([ATHLON-DeltaLon, ATHLON+DeltaLon, ATHLAT-DeltaLat, ATHLAT+DeltaLat], crs=ccrs.PlateCarree())

ax.set_title("Gridpoints and Boxes over Ireland for a 1° Forecast Resolution")

grd = ax.gridlines(draw_labels=True, linewidth=0)
grd.xlocator = plt.FixedLocator(range(int(ATHLON-DeltaLon), int(ATHLON+DeltaLon)))
grd.top_labels = False
grd.right_labels = False
plt.savefig('../Thesis/Gridpoints_stars.png')
```
